```python
g0to34 = df.loc[df['subgroup_cat1'] == '0-34']
g0to34['aaRate10K'] = g0to34['aaRate10K'].fillna((g0to34['count']/52979)*10000)
g0to34
```

```python
g35to54 = df.loc[df['subgroup_cat1'].isin(['35-44','45-54'])]
g35to54['aaRate10K'] = g35to54['aaRate10K'].fillna((g35to54['count']/136273)*10000)
g35to54
```

```python
g55to64 = df.loc[df['subgroup_cat1'] == '55-64']
g55to64['aaRate10K'] = g55to64['aaRate10K'].fillna((g55to64['count']/113362)*10000)
g55to64
```

```python
g65to74 = df.loc[df['subgroup_cat1'] == '65-74']
g65to74['aaRate10K'] = g65to74['aaRate10K'].fillna((g65to74['count']/191131)*10000)
g65to74
```

```python
g75 = df.loc[df['subgroup_cat1'].isin(['75-84', '85+'])]
g75['aaRate10K'] = g75['aaRate10K'].fillna((g75['count']/212179)*10000)
g75
```

```python
ageGroup = pd.concat([g0to34, g35to54, g55to64, g65to74,g75 ], axis=0)
ageGroup
```

```python
ageGrp_index = df[(df.subgroup1 == 'AgeGroup')&(df.subgroup_cat1 != '35+')].index
df.drop(index=ageGrp_index, inplace=True)
df
```

```python
new_data = pd.concat([df, ageGroup], axis=0)
new_data
```

```python
new_data[(new_data.subgroup1 == 'AgeGroup')&(new_data.subgroup_cat1 != '35+')]
```

```python
new_data.drop(columns=['indicator', 'aaRate10Kpy', 'cRate10Kpy'], inplace =True)
```

```python
new_data
```

```python
gTotal = new_data[new_data['Year'] == '2000-2018']
gTotal['subgroup2'] = 'Grand Total'
gTotal['subgroup_cat2'] = gTotal['subgroup_cat2'].fillna('Grand Total')
gTotal
```

```python
new_data.drop(index=new_data[new_data['Year'] == '2000-2018'].index, inplace=True)
new_data
```

```python
new_data = pd.concat([new_data,gTotal], axis = 0)
new_data
```

```python
new_data['subgroup_cat2'] = new_data['subgroup_cat2'].fillna(new_data['subgroup2'])
new_data
```
