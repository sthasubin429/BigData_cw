```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_json('income_NY_sample.json', orient='records')
df2 = pd.read_csv('hospitalization_clean.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df2.rename(columns = {'County':'county'}, inplace = True)
df2
```

```python
data = pd.merge(df1, df2, on='county', how='inner')
data
```

```python
data1 = pd.DataFrame(data.groupby(["income"])['dailyMean'].mean())
data1.plot.line(subplots=True, figsize=(10,10))
```

```python
data1 = pd.DataFrame(data[(data.county.isin(['Albany', 'Allegany', 'Bronx', 'Broome', 'Cattaraugus', 'Cayuga',
                                               'Chautauqua', 'Chemung', 'Chenango', 'Clinton', 'Columbia',
                                               'Queens', 'Rensselaer', 'Richmond', 'Rockland', 'St. Lawrence',]))]
             .groupby(["income"])[['count', 'aaRate10K','cRate10K']].mean())

data1.plot.line(subplots=True, figsize=(12,12))
```

```python
data2 = pd.DataFrame(data[(data.county.isin(['Albany', 'Allegany', 'Bronx', 'Broome', 'Cattaraugus', 'Cayuga',
                                               'Chautauqua', 'Chemung', 'Chenango', 'Clinton', 'Columbia',
                                               'Queens', 'Rensselaer', 'Richmond', 'Rockland', 'St. Lawrence',]))]
             .groupby(["ethnicity"])[['count', 'aaRate10K','cRate10K']].mean())

data2.plot.line(subplots=True, figsize=(12,12))
```

```python
data2 = pd.DataFrame(data[(data.county.isin(['Albany', 'Allegany', 'Bronx', 'Broome', 'Cattaraugus', 'Cayuga',
                                               'Chautauqua', 'Chemung', 'Chenango', 'Clinton', 'Columbia',
                                               'Queens', 'Rensselaer', 'Richmond', 'Rockland', 'St. Lawrence',]))]
             .groupby(["income_groups"])[['count', 'aaRate10K','cRate10K']].mean())

data2.plot.line(subplots=True, figsize=(12,12))
```
