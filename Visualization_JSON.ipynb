```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sample = pd.read_json('income_NY_sample.json', orient='records')
```

```python
sample
```

```python
sns.catplot(y="income", kind="count", palette="pastel", data=sample[sample["county"] == "Queens"],  
            height=6, aspect=10/6)
```

```python
sns.catplot(y="income_groups", kind="count", palette="pastel", data=sample[sample["county"] == "Queens"],  
            height=6, aspect=10/6)
```

```python
sns.catplot(y="education_level", kind="count", palette="pastel", data=sample[sample["county"] == "Queens"],  
            height=6, aspect=10/6)
```

```python
sns.catplot(y="development_region", hue="income", kind="count",
            palette="pastel", edgecolor=".6",
            data=sample,
            height=8.27, aspect=11.7/8.27)
```

```python
plt.figure(figsize=(20,15))
sns.catplot(y="county", hue="income", kind="count",
            palette="pastel", edgecolor=".6",
            data=sample[sample['county'].isin( ["Queens", "Kings", "Bronx", "New York"])], 
            height=8.27, aspect=11.7/8.27)
```

```python
sns.catplot(y="education_level", hue="income", kind="count",
            palette="pastel", edgecolor=".6",
            data=sample, 
            height=8.27, aspect=11.7/8.27)
```

```python
sns.catplot(y="ethnicity", hue="income", kind="count",
            palette="pastel", edgecolor=".6",
            data=sample, 
            height=8.27, aspect=11.7/8.27)
```

```python
data = sample.groupby("income")["_id"].count()
labels = data.keys()
data.plot.pie(autopct="%.1f%%", radius =2, labels=labels, explode=[0.05]*len(data), pctdistance=0.5)
plt.legend( bbox_to_anchor=(2,1), loc="upper left")
```

```python
data = dict(sample.groupby("income")["_id"].count())
data.values()
```

```python
data = sample.groupby("income_groups")["_id"].count()
labels = data.keys()
data.plot.pie(autopct="%.1f%%", radius =2, labels=labels, explode=[0.05]*len(data), pctdistance=0.5)
plt.legend( bbox_to_anchor=(2,1), loc="upper left")
```

```python
data = sample.groupby("ethnicity")["_id"].count()
labels = data.keys()
data.plot.pie(autopct="%.1f%%", radius =2, labels=labels, explode=[0.05]*len(data), pctdistance=0.5)
plt.legend( bbox_to_anchor=(2,1), loc="upper left")
```

```python
sns.catplot(y="households_with_elderly", kind="count", palette="pastel", data=sample , height=3,)
sns.catplot(y="households_with_children", kind="count", palette="pastel", data=sample,  height=3, )
```
